```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('harmonic_oscillator_no_temp_k_10_to_16_2_step.json') as f:
    data = json.load(f)
```

```python
energies = {int(key):[sets[0] for sets in value] for key, value in data.items() }
indicator = {int(key):[sets[1] for sets in value] for key, value in data.items() }
tstar = {int(key):[sets[2] for sets in value] for key, value in data.items() }
eigenvalue = {int(key):[sets[3] for sets in value] for key, value in data.items() }

line_formats={10:'-',12:'--',14:'-.',16:':'}
```

```python
for k in energies.keys():
    plt.plot(energies[k],indicator[k],linestyle=line_formats[k],label=f"K={k}")
plt.title(f"Allowed Energies for the Harmonic Oscillator",fontsize=16)
plt.xlabel("Energy (Natural Units)",fontsize=14)
plt.ylabel("Indiactor Function",fontsize=14)
plt.grid(which="both",alpha=0.3)
plt.legend(fontsize=14)
plt.show()
```

```python
plt.plot(energies[16],1+np.array(tstar[16]),label=r"$t_\star$")
plt.plot(energies[16],1+np.array(eigenvalue[16]),linestyle="--",label=r"$\text{Min}(\text{Spec}(M^{(16)}(E)))$")
plt.axvline(x=0.5,linestyle=":",color="r",label="Analytic Spectrum")
plt.axvline(x=1.5,linestyle=":",color="r")
plt.axvline(x=2.5,linestyle=":",color="r")
plt.axvline(x=3.5,linestyle=":",color="r")
plt.axvline(x=4.5,linestyle=":",color="r")
plt.title(r"Comparison of $t_\star$ with the Eigenvalues of the moment matrix",fontsize=16)
plt.xlabel("Energy (Natural Units)",fontsize=14)
plt.ylabel("log("+r"$\text{Min}(\text{Spec}(M^{(16)}(E)))$"+") or log("+r"$t_\star$"+")",fontsize=14)
plt.xlim(0,5)
plt.yscale("log")
plt.grid(which="both",alpha=0.3)
plt.legend(fontsize=14)
plt.show()
```
